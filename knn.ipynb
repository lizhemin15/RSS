```python
# Tucker for random sampling
import rss

parameters = {}
net_list = []
net_list.append({'net_name':'TF','sizes':[10,10],'dim_cor':[10,10],'mode':'tucker'})

parameters['net_p'] = {'gpu_id':'cpu','net_name':'composition','net_list':net_list}
parameters['data_p'] = {'data_shape':(10,10),'random_rate':0.5,
                        'pre_full':True,'mask_type':'down_sample','down_sample_rate':4}
parameters['train_p'] = {'train_epoch':20}
parameters['show_p'] = {'show_type':'gray_img','show_content':'original'}
rssnet = rss.rssnet(parameters)

rssnet.show()
rssnet.show_p['show_content'] = 'recovered'
for i in range(10):
    rssnet.train()
    rssnet.show()
```
